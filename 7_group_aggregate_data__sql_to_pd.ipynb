## 7.&nbsp;Add the average price of each publisher - book type combination from your previous query

````python
titles.head()
````

````python
titles_original = titles.copy() # Make a copy of the original DataFrame before any aggregations
titles = titles_original.copy() # Assuming 'titles_original' is your original DataFrame before any aggregations
titles.head()
````

````python
Avg_price = titles.groupby(['pub_id', 'type'])['price'].mean().reset_index()
cc_price = titles.groupby(['pub_id', 'type'])['price'].count().reset_index()

# Merge the average price and count of prices back to the 'titles' DataFrame
titles = titles.merge(Avg_price, on=['pub_id', 'type'],suffixes=('', '_avg'))
titles = titles.merge(cc_price, on=['pub_id', 'type'],suffixes=('', '_count'))

# Rename the columns to 'Avg_price' and 'cc_price'
#titles.rename(columns={'price_x': 'Avg_price', 'price_x': 'cc_price'}, inplace=True)

# Select the desired columns
titles.loc[:, ['pub_id', 'type', 'price_avg', 'price_count']]
````

### Hint:

In SQL the syntax is:

```sql
SELECT  AVG(price), COUNT(*), pub_id,type
FROM titles
GROUP BY pub_id, type;
```

## 8.&nbsp;From your previous query, keep only the combinations of publisher - book type with an average price higher than 12

````python
#need to return to the original data and remove the former aggregation and then apply the new functions, otherwise the price_avg and price_count can not be found at all
titles = pd.read_csv(gd_path(files_id["titles"]), sep=";")
titles_original = titles.copy() # Make a copy of the original DataFrame before any aggregations
titles = titles_original.copy() # Assuming 'titles_original' is your original DataFrame before any aggregations
titles.head()
````

````python
Avg_price = titles.groupby(['pub_id', 'type'])['price'].mean().reset_index()
cc_price = titles.groupby(['pub_id', 'type'])['price'].count().reset_index()

# Merge the average price and count of prices back to the 'titles' DataFrame
titles = titles.merge(Avg_price, on=['pub_id', 'type'],suffixes=('', '_avg'))
titles = titles.merge(cc_price, on=['pub_id', 'type'],suffixes=('', '_count'))

# Rename the columns to 'Avg_price' and 'cc_price'
#titles.rename(columns={'price_x': 'Avg_price', 'price_x': 'cc_price'}, inplace=True)

# Select the desired columns with condition
titles.loc[titles['price_avg']>12, ['pub_id', 'type', 'price_avg', 'price_count']]
````

### Hint:

In SQL the syntax is:

```sql
SELECT pub_id,type, COUNT(*)
FROM titles
GROUP BY pub_id, type
HAVING AVG(price) > 12;
```

## 9.&nbsp;Order the results of your previous query by these two criteria:
1. Count of books, descendingly
2. Average price, descendingly

````python
#need to return to the original data and remove the former aggregation and then apply the new functions, otherwise the price_avg and price_count can not be found at all
titles = pd.read_csv(gd_path(files_id["titles"]), sep=";")
titles_original = titles.copy() # Make a copy of the original DataFrame before any aggregations
titles = titles_original.copy() # Assuming 'titles_original' is your original DataFrame before any aggregations
titles.head()
````

````python
Avg_price = titles.groupby(['pub_id', 'type'])['price'].mean().reset_index()
cc_price = titles.groupby(['pub_id', 'type'])['price'].count().reset_index()

# Merge the average price and count of prices back to the 'titles' DataFrame
titles = titles.merge(Avg_price, on=['pub_id', 'type'],suffixes=('', '_avg'))
titles = titles.merge(cc_price, on=['pub_id', 'type'],suffixes=('', '_count'))

# Rename the columns to 'Avg_price' and 'cc_price'
#titles.rename(columns={'price_x': 'Avg_price', 'price_x': 'cc_price'}, inplace=True)

# Select the desired columns with condition
filt_titles = titles.loc[titles['price_avg']>12, ['pub_id', 'type', 'price_avg', 'price_count']]
filt_titles = filt_titles.sort_values(by = ['price_count','price_avg'], ascending=[False,False])
filt_titles.head()
````

### Hint:

In SQL the syntax is:

```sql
SELECT  pub_id, type, COUNT(*), AVG(price)
FROM titles
GROUP BY pub_id, type
HAVING AVG(price) > 12
ORDER BY COUNT(*) DESC, AVG(price) DESC;
```

## 10.&nbsp;Some authors have a contract, while others don't - it's indicated in the "contract" column of the authors table.
     
    Select all the states and cities where there are 2 or more contracts overall

````python
authors.head()
````

````python
#need to return to the original data and remove the former aggregation and then apply the new functions, otherwise the price_avg and price_count can not be found at all
authors = pd.read_csv(gd_path(files_id["authors"]), sep=";")
authors_original = authors.copy() # Make a copy of the original DataFrame before any aggregations
authors = authors_original.copy() # Assuming 'titles_original' is your original DataFrame before any aggregations
authors.head()
````

````python
sum_cont = authors.groupby(['state','city'])['contract'].sum().reset_index()
authors = authors.merge(sum_cont, on=['state', 'city'],suffixes=('', '_sum'))
filt_authors = authors.loc[authors['contract_sum']>1, ['state', 'city', 'contract_sum']]
filt_authors.head(20)
````

### Hint:

In SQL the syntax is:

```sql
SELECT SUM(contract),city,state
FROM authors
GROUP BY state,city
HAVING SUM(contract)>1;
```
