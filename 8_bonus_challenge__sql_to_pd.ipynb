# From SQL to pandas challenge 8

````python
# import libraries
import pandas as pd

# load data
# This code is made to load our data stored on Google Drive
def gd_path(file_id):
    """Generate a shareable link from Google Drive file id."""
    return f"https://drive.google.com/uc?export=download&id={file_id}"

# Google Drive file ids
files_id = {
    "sales": "1fzFc9rwYmVIPaGOFmhLVxCi3kg19vNU2",
    "authors": "1fEF89Nhe61EebAljKlwFwfEuokK0o6aJ"
}

# Read data from Google Drive
sales = pd.read_csv(gd_path(files_id["sales"]), sep=";")
authors = pd.read_csv(gd_path(files_id["authors"]), sep=";")
````

## 1.&nbsp;Select the top 5 sales with most quantity sold between 1993-03-11 and 1994-09-13 from the table sales

````python
sales.head()
````

````python
df = pd.DataFrame(sales)
print (df)
print (df.dtypes)
````

````python
# Notice that the ‘ord_date’ were indeed stored as strings (represented by object).
# convert the strings to datetime in Pandas DataFrame
df['ord_date'] = pd.to_datetime(df['ord_date'])
filt = sales.loc[(sales['ord_date'] >= '1993-03-11') & (sales['ord_date'] <= '1994-09-13')]
filt_sales = filt.sort_values(by = 'qty', ascending=False)
filt_sales.head(5)
````

### Hint:

In SQL the syntax is:

```sql
SELECT *
FROM sales
WHERE ord_date BETWEEN '1993-03-11' AND '1994-09-13'
ORDER BY qty DESC
LIMIT 5;
```

## 2.&nbsp;How many authors have an "i" in their first name, are from Utah, Maryland, or Kansas?

````python
authors.head()
````

````python
df = pd.DataFrame(authors)
filt = df['au_fname'].str.contains('i', regex=False)
filt_authors = df.loc[filt,['state']].isin(['CA','MD','KS'])
#filt_authors[['au_fname','state']] # what is wrong, if i want to check the relavant columns
print(filt_authors.count())
````

### Hint:

In SQL the syntax is:

```sql
SELECT COUNT(*)
FROM authors
WHERE au_fname LIKE '%i%'
AND state IN ('CA' , 'MD', 'KS');
```

## 3.&nbsp;In California, how many authors are there in cities that contain an "o" in the name?
   - Show only results for cities with more than 1 author.
   - Sort the cities ascendingly by author count.

````python
df = pd.DataFrame(authors)
df.head()
````

````python
filt = df['city'].str.contains('o', case=False, regex=False)
filt_authors = df.loc[(filt) & (df['state'] == 'CA')]
city_author_count = filt_authors.groupby('city')['au_fname'].count()
filt_fname = city_author_count[city_author_count > 1].sort_values(ascending=False)
print(filt_fname)
````

### Hint:

In SQL the syntax is:

```sql
SELECT city, COUNT(au_fname)
FROM authors
WHERE state = 'CA'
AND city LIKE '%o%'
GROUP BY city
HAVING COUNT(au_fname) > 1
ORDER BY COUNT(au_fname);
```
