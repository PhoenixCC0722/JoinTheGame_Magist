```python
category_type_df_1st = category_type_df.loc[category_type_df["type"] == "11865403"]
```

```python
import nltk
nltk.download('punkt')

top_N = 5
# replace all non-alphanumeric characters
category_type_df_1st["name_rep"] = category_type_df_1st.name.str.lower().str.replace('\W', ' ', regex=True)

# tokenize, Tokenizing the words, splits each string into a list.
# In this example, looking at df will show a tok column, where each row is a list
category_type_df_1st['tok'] = category_type_df_1st.name_rep.apply(nltk.tokenize.word_tokenize)

# all tokenized words to a list
words = category_type_df_1st.tok.tolist()  # this is a list of lists
words = [word for list_ in words for word in list_]

# frequency distribution
word_dist = nltk.FreqDist(words)

# remove stopwords
# stopwords = nltk.corpus.stopwords.words('english')
# words_except_stop_dist = nltk.FreqDist(w for w in words if w not in stopwords)

# output the results
rslt = pd.DataFrame(word_dist.most_common(top_N), columns=['Word', 'Frequency'])
print(rslt)
```

```python
# Get the top 3 most common words and their frequencies
top_3_words = word_dist.most_common(3)

# Extract the words from the tuples
top_3_words_str = [word for word, freq in top_3_words]

# Concatenate the words
category_1st = ' '.join(top_3_words_str)
print(category_1st)
```

## Now, we should loop the finding Keywords function inside of the top 30 types, in each time, we go throw in the 'name' column and find out the most frequent words and regard it as the category name. Then put it into the new column 'category_name'

How to realize this function?

```python
#	type  a numerical code for product type
category_top30_types_df = category_type_df.groupby("type").count().nlargest(30, "sku").reset_index()
category_top30_types_df.info()
category_top30_types_df.head()
```

```python
category_name = {}
top_N = 3
for i in range(len(category_top30_types_df)):
    num = category_top30_types_df.type.iloc[i]
    category_type_df_i = category_type_df.loc[category_type_df["type"] == num]
    category_type_df_i_copy = category_type_df_i.copy()
    category_type_df_i_copy["name_rep"] = category_type_df_i_copy.name.str.lower().str.replace('\W', ' ', regex=True)
    category_type_df_i_copy['tok'] = category_type_df_i_copy.name_rep.apply(nltk.tokenize.word_tokenize)
    words = category_type_df_i_copy.tok.tolist()  # this is a list of lists
    words = [word for sublist in words for word in sublist]  # flatten the list of lists
    word_dist = nltk.FreqDist(words)
    rslt = pd.DataFrame(word_dist.most_common(top_N), columns=['Word', 'Frequency'])
    top_3_words = word_dist.most_common(3)
    top_3_words_str = [word for word, freq in top_3_words]
    category_i = ' '.join(top_3_words_str)
    # Store results for each category separately
    category_name[num] = {'rslt': rslt, 'names': category_i} # dictionary
```

```python

# Initialize an empty list to store data for each category
data = []

# Iterate through each category in the category_name dictionary
for category_id, values in category_name.items():
    category_data = {}
    category_data['Type'] = category_id  # Assign the category ID

    # Extract the 'rslt' and 'names' values from the dictionary
    rslt_df = values['rslt']
    category_data['Word'] = ', '.join(rslt_df['Word'].astype(str).tolist())  # Convert Word column to comma-separated string
    category_data['Frequency'] = ', '.join(rslt_df['Frequency'].astype(str).tolist())  # Convert Frequency column to comma-separated string
    category_data['Names'] = values['names']  # Assign the 'names' value

    # Append the category data to the list
    data.append(category_data)

# Create a DataFrame from the list of category data
result_df = pd.DataFrame(data)
# Display the DataFrame
#print(result_df)
```
